```python
from geotext import GeoText
import numpy as np
import pandas as pd
import re

import spacy

nlp = spacy.load("en_core_web_sm")

def get_cities_countries(c_list):
    '''
    return a string of comma separated of list
    '''
    cc_l = np.nan
    if len(c_list) > 0:
        cc_l = ''
        for c_ in c_list:
            cc_l +=c_+','
        cc_l = cc_l.strip(',')
    return cc_l

def get_df_countries_cities(filename):
    
    all_list = []

    file = open(filename, 'r')
    for line in file.readlines():
        line = line.strip()
        
        entities = nlp(line)
        
        cities_countries = [ent.text for ent in entities.ents if ent.label_=='GPE']
        print(cities_countries)
        countries = []
        cities = []
        for ci_cou in cities_countries:
            places = GeoText(ci_cou)
            
            if len(places.countries)>0:
                countries.append(ci_cou)
                
            if len(places.cities)>0:
                cities.append(ci_cou)
            
        

        countries = get_cities_countries(countries)

        cities = get_cities_countries(cities)



        all_list.append({'headline': line, 'countries':countries, 'cities': cities})



    file.close()

    return pd.DataFrame(all_list)

get_df_countries_cities("headlines.txt")
```
